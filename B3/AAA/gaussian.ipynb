```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros de las 3 gaussianas
means = [0, 5, 10]       # Medias
std_devs = [1, 1.5, 1]   # Desviaciones estndar
weights = [0.4, 0.4, 0.2]  # Pesos (deben sumar 1)

# Nmero de puntos
n_points = 1000

# Generar los puntos para cada gaussiana segn su peso
n_points_gaussian1 = int(weights[0] * n_points)
n_points_gaussian2 = int(weights[1] * n_points)
n_points_gaussian3 = int(weights[2] * n_points)

# Generar los datos
data_gaussian1 = np.random.normal(means[0], std_devs[0], n_points_gaussian1)
data_gaussian2 = np.random.normal(means[1], std_devs[1], n_points_gaussian2)
data_gaussian3 = np.random.normal(means[2], std_devs[2], n_points_gaussian3)

# Concatenar todos los datos
data = np.concatenate([data_gaussian1, data_gaussian2, data_gaussian3])

# Guardar los datos en un fichero
# np.savetxt('datos_mixtura_gaussianas.txt', data)

# Mostrar un histograma de los datos
plt.hist(data, bins=50, density=True, alpha=0.6, color='palegreen', label='Datos generados')

# Calcular y mostrar la funcin de la mixtura de las 3 gaussianas
x = np.linspace(min(data), max(data), 1000)
mixture_pdf = (weights[0] * (1/(std_devs[0] * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - means[0])/std_devs[0])**2) +
               weights[1] * (1/(std_devs[1] * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - means[1])/std_devs[1])**2) +
               weights[2] * (1/(std_devs[2] * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - means[2])/std_devs[2])**2))

plt.plot(x, mixture_pdf, color='red', label='Mixtura de Gaussianas', linewidth=1)

# Opcional: Mostrar las gaussianas individuales para comparacin
y1 = weights[0] * (1/(std_devs[0] * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - means[0])/std_devs[0])**2)
y2 = weights[1] * (1/(std_devs[1] * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - means[1])/std_devs[1])**2)
y3 = weights[2] * (1/(std_devs[2] * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - means[2])/std_devs[2])**2)

plt.plot(x, y1, color='black', linestyle='--', label='Gaussiana 1', linewidth=0.6)
plt.plot(x, y2, color='black', linestyle='--', label='Gaussiana 2', linewidth=0.6)
plt.plot(x, y3, color='black', linestyle='--', label='Gaussiana 3', linewidth=0.6)

plt.title('Mixtura de 3 Gaussianas')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
# plt.legend()
# plt.savefig('mixtura_gaussianas_original.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros de las 3 gaussianas
means = [-5, 7]      # Medias
std_devs = [4, 4]    # Desviaciones estndar
weights = [0.6, 0.4] # Pesos (deben sumar 1)

# Nmero de puntos
n_points = 1000

# Generar los puntos para cada gaussiana segn su peso
n_points_gaussian1 = int(weights[0] * n_points)
n_points_gaussian2 = int(weights[1] * n_points)

# Generar los datos
np.random.seed(23)
data_gaussian1 = np.random.normal(means[0], std_devs[0], n_points_gaussian1)
data_gaussian2 = np.random.normal(means[1], std_devs[1], n_points_gaussian2)

# Concatenar todos los datos
data = np.concatenate([data_gaussian1, data_gaussian2])

# Guardar los datos en un fichero
np.savetxt('data_question4.txt', data)

# Tamao de la grfica
plt.figure(figsize=(8, 4))

# Mostrar un histograma de los datos
plt.hist(data, bins=50, density=True, alpha=0.6, color='palegreen', label='Datos generados', edgecolor='black', linewidth=0.5, width=0.5)

# Calcular y mostrar la funcin de la mixtura de las 3 gaussianas
x = np.linspace(min(data), max(data), 1000)
mixture_pdf = (weights[0] * (1/(std_devs[0] * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - means[0])/std_devs[0])**2) +
               weights[1] * (1/(std_devs[1] * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - means[1])/std_devs[1])**2))

plt.plot(x, mixture_pdf, color='red', label='Mixtura de Gaussianas', linewidth=1)

# Mostrar las gaussianas individuales para comparacin
y1 = weights[0] * (1/(std_devs[0] * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - means[0])/std_devs[0])**2)
y2 = weights[1] * (1/(std_devs[1] * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - means[1])/std_devs[1])**2)

plt.plot(x, y1, color='black', linestyle='--', label='Gaussiana 1', linewidth=0.6)
plt.text(-4, 0.06, '', fontsize=12)
plt.plot(x, y2, color='black', linestyle='--', label='Gaussiana 2', linewidth=0.6)
plt.text(12, 0.02, '', fontsize=12)

# Tamao de los valores de los ejes
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# plt.xlabel('Valor', fontsize=10)
# plt.ylabel('Densidad', fontsize=10)
plt.savefig('mixtura_gaussianas_original.png', dpi=300)
plt.show()
```
